```python
cdf = df[['Area', 'Room', 'Parking', 'Warehouse', 'Elevator', 'Price']]
cdf.head(9)
```

#### - Normization and Preprocessing

```python
cdf['Area'] = pd.to_numeric(cdf['Area'], errors='coerce')
cdf['Price'] = pd.to_numeric(cdf['Price'], errors='coerce')


cdf['Parking'] = cdf['Parking'].astype(int)
cdf['Warehouse'] = cdf['Warehouse'].astype(int)
cdf['Elevator'] = cdf['Elevator'].astype(int)
```

## 5- Split data into Train & Test

```python
msk = np.random.rand(len(df)) < 0.8
train = cdf[msk]
test = cdf[~msk]

print(msk)
print(~msk)
print("--------------------------- cdf -------------------------------")
print(cdf)
print("-------------------------- Train ---------------------------------")
print(train)
print("-------------------------- Test ----------------------------------")
print(test)
```

## 6- Plotting Regression Model

### Train & Test Data Distribution

- Area

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)

ax1.scatter(train.Area, train.Price, color = 'blue')
ax1.scatter(test.Area, test.Price, color = 'red')

plt.xlabel("Area")
plt.ylabel("Price")
plt.show()
```

- Room

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)

ax1.scatter(train.Room, train.Price, color = 'blue')
ax1.scatter(test.Room, test.Price, color = 'red')

plt.xlabel("Room")
plt.ylabel("Price")
plt.show()
```

## 7- Linear Regression Model

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()

train = train.dropna(subset=['Area','Room','Price'])

x = np.asanyarray(train[['Area','Room']])
y = np.asanyarray(train[['Price']])
reg.fit(x, y)

# The Coefficients
print(reg)
print('Coefficients: ', reg.coef_)
print('Intercept: ', reg.intercept_)
```

```python
print(test.isnull().sum())
```

```python
test = test.dropna(subset=['Area', 'Room', 'Price'])
```

## 7- Evaluation Model

```python
y_hat = reg.predict(test[['Area','Room']])
x = np.asanyarray(test[['Area','Room']])
y = np.asanyarray(test[['Price']])

print("Residental Sum of Squares: %.2f" % np.mean((y_hat - y) ** 2))
print("Variance Score: %.2f" % reg.score(x, y))
```
