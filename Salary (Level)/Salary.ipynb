### Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Loading Dataset

```python
dataset = pd.read_csv('salary.csv')
x = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

## 1- Training the Linear Regression model

```python
from sklearn.linear_model import LinearRegression
linear_regressor = LinearRegression()
linear_regressor.fit(x, y)
```

### Linear Regression Visualization

```python
plt.scatter(x, y, color = "red")
plt.plot(x, linear_regressor.predict(x), color = "blue")
plt.title('Linear Regression')
plt.ylabel('salary')
plt.xlabel('Level')
plt.show()
```

### Checking New Data

```python
linear_regressor.predict([[6.5]])
```

## 2- Training the Polynomial Regresion model

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2)
x_poly = poly.fit_transform(x)
linear_regressor2 = LinearRegression()
linear_regressor2.fit(x_poly, y)

# x
# x_poly
```

### Polynomial Regresion Visualization

```python
plt.scatter(x, y, color = 'red')
plt.plot(x, linear_regressor2.predict(poly.fit_transform(x)), color = 'blue')
plt.title('Polynomial Regression')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```

### Checking New Data

```python
linear_regressor2.predict(poly.fit_transform([[6.5]]))
```
